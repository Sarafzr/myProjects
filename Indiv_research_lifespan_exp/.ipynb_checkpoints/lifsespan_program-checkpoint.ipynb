```python
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
dataset = pd.read_excel("lifespan_data.xlsx")
```

```python
dataset.head()
```

# Lineplot to show time series data between days and alive

```python
sb.lineplot(x='DAY', y='A', data=dataset, hue="STRAIN")
```

```python
#new data frame to find when 'A' (Alive) is equal to zero
df_lastday=dataset[dataset['A']==0]
display(df_lastday)
```

```python
#new data frames to calculate min and mean

#new data frame to group by strain and day to then find the minimum day to know when the starin first died
grouped_min = df_lastday.groupby('STRAIN')['DAY'].min().reset_index()

#created a new column with the data frame grouped_min
grouped_min["strain_type"]= ["Alpha", "Alpha", "Alpha", "HB", "HB", "HB"]

#created a new data frame called group_mean with the grouped.min data to group by the strain type and day to calculate the mean
grouped_mean=grouped_min.groupby("strain_type")['DAY'].mean().reset_index()
```

```python
display(grouped_min)
display(grouped_mean)
```

# Barplot to show the average lifespan for the Alpha strains and the HB strains

```python
sb.barplot(x='strain_type', y='DAY', data=grouped_mean, orient="v")
```
